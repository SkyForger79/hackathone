```python
import numpy as np
import pandas as pd 
from keras.preprocessing.image import ImageDataGenerator, load_img
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import random
import os
```

```python
x_train = np.load('../input/eyes-dataset/x_train.npy').astype(np.float32)
y_train = np.load('../input/eyes-dataset/y_train.npy').astype(np.float32)
x_val = np.load('../input/eyes-dataset/x_val.npy').astype(np.float32)
y_val = np.load('../input/eyes-dataset/y_val.npy').astype(np.float32)
```

```python
train_datagen = ImageDataGenerator(rescale=1./255, rotation_range=10, width_shift_range=0.2, height_shift_range=0.2, shear_range=0.2)
train_generator = train_datagen.flow(x=x_train, y=y_train, batch_size=32, shuffle=True)

val_datagen = ImageDataGenerator(rescale=1./255)
val_generator = val_datagen.flow(x=x_val, y=y_val, batch_size=32, shuffle=False)
```

```python
from keras.models import Sequential
from keras import layers

model = Sequential()

model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(26, 34, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Flatten())
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
history = model.fit_generator(train_generator, epochs=8, validation_data=val_generator)
```

```python
predict = model.predict_generator(val_generator)
model.save("model7.h5")
```
